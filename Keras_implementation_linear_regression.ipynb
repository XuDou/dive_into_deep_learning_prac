```
import tensorflow as tf
from tensorflow import keras
```

```
import random 
import numpy as np
import matplotlib.pyplot as plt
```

```
# generate the simulation dataset

true_w = true_w = np.array([[2], [-1.6]])
true_b = 4.2  
X = np.random.normal(loc=0.0,scale=1.0,size=(3000,2))
epsilon = np.random.normal(loc=0.0,scale=0.01,size=(3000,1))
label_y = np.matmul(X,true_w) + true_b + epsilon
```

```
f, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,sharey=True,constrained_layout=True,\
                            figsize=(10,4))  #plt.subplot(121)
plt.figure(dpi=3000)

ax1.scatter(X[:,0],label_y,2)
ax1.set_xlabel('X_0',fontsize=12)
ax1.set_ylabel('label_y',fontsize=15)




ax2.scatter(X[:,1],label_y,2)
ax2.set_xlabel('X_1',fontsize=12)
```

```
# build the linear regression model
linear_model = keras.models.Sequential([keras.layers.Dense(1)]) 
```

```
from sklearn.model_selection import train_test_split

X_train_full, X_test, y_train_full, y_test = train_test_split(X, label_y)

X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full)
```

```
linear_model.compile(loss="mean_squared_error",optimizer="sgd")
history = linear_model.fit(X_train, y_train, epochs=30, validation_data=(X_valid, y_valid))
```
